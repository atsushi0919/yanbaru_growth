### 1. 概要

#### 模写コーディング挑戦前の状態

- HTML+CSS初心者（CSSの存在は知っているが書けない）
- 「bootstrap」を少しやってみたけど、いまいち理解できてない

案件に関わるとなるとフロント技術が必要と思い、メンターさんに相談したところ、模写コーディングが良いと聞きました。  
5件くらい模写すればある程度のスキルが身につくのでは？との事でした。

最初は何をするにも時間がかかりまくって苦しかったけど、とりあえず4件ほど模写してみたら・・・

- 多少複雑でも動きがないページなら模写出来るようになった
- 動きがある要素も調べながら作れるようになった

という収穫があったので __「模写コーディングをやってみたお話」__ をします。

---

### 2. 模写方針

初心者がいきなり完璧な模写を狙うと心が折れるので妥協点を事前に設定しておきます。（自分ルールの設定）

- フォントは気にしない
- 色は気にしない
- 画像は気にしない
- 細かい余白は気にしない
- 動作はしっかり模写する
  - マウスオーバー
  - レスポンシブ対応
- 定規ツールやスポイトツールの使用可
- 入門書やWeb情報をググるのはOK
- ソースはどうしてもわからなかったときだけ見る

---

模写コーディングの進め方などについて詳しく書いてあったので、模写コーディング練習サイト「Codestep」を利用することにしました。

### 4. 模写作業

1.サイト全体を把握する
  - 全体のレイアウトやページリンク
  - 要素に動きがあるものを確認
  - 画面サイズが変わった時の表示

2.ページ構成を書き出す
  - デザインツール(XD, figma, sketchなど)やノートにページ構成をまとめる  

3.HTML＋CSS模写
  - 全体的にざっくりと作ってから細部を調整する

4.動きがある要素を作る
  - 動きがある要素の動作を作る

5.レスポンシブ対応
  - 画面サイズが変わった時の動きを作る

6.答え合わせ
  - 模写したサイトのソースと自分のソースを見比べてみる  
  - 動きのある要素やブレークポイントなどを確認する  
  - 可能ならタブレットやスマホ実機などで確認してみる  

---

### 5. 4件模写してみた結果

入門: 初めてなので何をするにも時間がかかった。4日で完成  
初級: 3日で完成  
中級1: マウスオーバー時の動きがどうしても分からず降参・カンニングした。7日で完成  
中級2: 「パララックス」をググって実装。5日 + スマホで見たときにリンクが正しく動作しない不具合修正1日

- 模写コーディングを4件やってみたら、ある程度コーディング出来るようになった！  
- 模写コーディングの後、ProgateのHTMLコース(練習・道場)をやってみたら余裕だった！
- 「動きのある要素」「スピードアップ」「精度向上」など、まだまだ勉強する必要あり！
