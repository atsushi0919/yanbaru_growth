### 1. Canvasとは

#### HTML5から追加されたHTMLの要素

- CanvasAPIの技術仕様に沿ってJavaScript等から指示を与えることで「図形や文字を描画」することができる
- HTMLでは `<canvas></canvas>` のように記述する
- 図形や文字の変形や色やスタイルの変更を「動的」に扱うことが出来る
- 「コンテキストオブジェクト(context object)」によって描画機能が提供される
- 図形を表示するには、canvasタグをサポートしたブラウザが必要 (余程古くなければ大丈夫)

---

### 2. コンテキストとは

- コンテキストとは、描画に関わる設定情報の状態（例えばペンの太さや色）
- コンテキストオブジェクトとは、コンテキストが構造化されたオブジェクト
- 「2Dのコンテキスト」と「3Dのコンテキスト」があり、総称して「描画コンテキスト」と呼ばれる

### 3. サンプルコード紹介

今回は、CanvasRenderringContext2Dオブジェクト（通称: Canvas2Dコンテキスト）を使用したアナログ時計の描画デモをご紹介します。

Canvas2Dコンテキストの主なプロパティ
    
| プロパティ | 説明 |
| :---: | :---: |
| context.strokeStyle | 線や輪郭の色 |
| context.fillStyle | 塗りつぶしの色 |
| context.lineWidth | 線の幅 |
| context.shadowColor | 影の色 |
| context.shadowBlur | 影のぼかし範囲 |

参考： [Canvasリファレンス](http://www.htmq.com/canvas/)
